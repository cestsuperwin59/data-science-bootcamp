```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file

df['ORD_YEAR'] = pd.to_datetime(df['ORD']).dt.strftime('%Y')

filter_cali_tex17 = df[((df['State'] == 'California') | (df['State'] == 'Texas')) & (df['ORD_YEAR'] == '2017')]
filter_cali_tex17.to_csv('filter_cali_tex17.csv')

filter_cali_tex17
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017

df_2017 = df[df['ORD_YEAR'] == '2017']
order_2017 = df_2017['Sales'].agg(['sum','std','mean'])
order_2017
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df_2018 = df[df['ORD_YEAR'] == '2018']

check_profit_2018 = df_2018[['Segment', 'Profit']].groupby('Segment').sum().reset_index()

check_profit_2018.head(1)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019

df_apr_dec2019 = df[(df['ORD'] >= '2019-04-15') & (df['ORD'] <= '2019-12-31')]
leastsales_states_2019 = df_apr_dec2019[['State', 'Sales']].groupby('State').sum().sort_values('Sales', ascending=True)
print(leastsales_states_2019.head(5).reset_index())
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
df_2019 = df[df['ORD_YEAR'] == '2019']

result_2019 = df_2019['Sales'].sum()
wc_sales = df_2019[(df_2019['Region'] == 'West') | (df_2019['Region'] == 'Central')]['Sales'].sum()

proportion = (wc_sales/result_2019*100).round(2)

print(f"West + Central Sales  {proportion}%  in 2019")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

df_2019_20 = df[(df['ORD_YEAR'] == '2019') | (df['ORD_YEAR'] == '2020')].reset_index()

top10_sales = df_2019_20.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).reset_index().head(10)
top10_order = df_2019_20.groupby('Product Name')['Order ID'].count().sort_values(ascending=False).reset_index().head(10)


print(top10_sales)
print(top10_order)
```
